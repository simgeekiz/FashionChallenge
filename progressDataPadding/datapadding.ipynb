## Estimating the max size of images

```python
import os
path, dirs, files = next(os.walk("images"))
file_count = len(files)
```

```python
image_sizes = []
for i in files:
    im_pth = "images/" + i
    im = Image.open(im_pth)
    image_sizes.append(im.size)
```

```python
# image_sizes
```

```python
desired_size = [max(image_sizes,key=lambda item:item[1])[1], max(image_sizes,key=lambda item:item[0])[0]]
desired_size
```

## Add Padding

```python
padded_images = []
```

```python
desired_size = 600
```

```python
index=0
for i in files:
    im_pth = "images/" + i
    im = Image.open(im_pth)
    old_size = im.size
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = im.resize(new_size, Image.ANTIALIAS)
    new_ima = Image.new("RGB", (desired_size,desired_size), color=(255,255,255))
    new_ima.paste(im, ((desired_size-new_size[0])//2,
                    (desired_size-new_size[1])//2))
    padded_images.append(new_ima)
    index = index + 1
```
